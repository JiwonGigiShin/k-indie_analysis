```python
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def artist_scraping(year_):
    
    #page 1
    url = f'https://koreanmusicawards.com/archive/?_sft_nominee_year={year_}'
    request = requests.get(url)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')

    artists = str(soup.select('a > div.nominee-item-title > div.nominee-artist'))
    artists2 = re.sub('<.+?>', '', artists, 0)
    artists2 = artists2.replace('[','').replace('&amp;', "&").replace(']','')
    artists_lst1 = artists2.split(", ")


    #page 2
    url = f'https://koreanmusicawards.com/archive/?_sft_nominee_year={year_}&sf_paged=2'
    request = requests.get(url)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')


    artists = str(soup.select('a > div.nominee-item-title > div.nominee-artist'))
    artists2 = re.sub('<.+?>', '', artists, 0)
    artists2 = artists2.replace('[','').replace('&amp;', "&").replace(']','')
    artists_lst2 = artists2.split(", ")



    #page 3
    url = f'https://koreanmusicawards.com/archive/?_sft_nominee_year={year_}&sf_paged=3'
    request = requests.get(url)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')

    artists = str(soup.select('a > div.nominee-item-title > div.nominee-artist'))
    artists2 = re.sub('<.+?>', '', artists, 0)
    artists2 = artists2.replace('[','').replace('&amp;', "&").replace(']','')
    artists_lst3 = artists2.split(", ")

    artist_all = []
    
    for a in artists_lst1:
        artist_all.append(a)

    for a in artists_lst2:
        artist_all.append(a)

    for a in artists_lst3:
        artist_all.append(a)
    
    
    return artist_all
```

```python
kma_artist_lst = []

for i in range(2008,2023):
    kma_artist_lst.append(artist_scraping(i))
```
